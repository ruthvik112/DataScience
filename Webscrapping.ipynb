# Getting Social Links From https://keithgalli.github.io/web-scraping/webpage.html

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r=requests.get("https://keithgalli.github.io/web-scraping/webpage.html")
webpage=bs(r.content)
print(webpage.prettify())
```

```python
social=webpage.select("body ul.socials li a ")
for i in range(len(social)):
    print(social[i]['href'])
```

# Scraping Table


```python
import pandas as pd

table=webpage.select("table.hockey-stats")[0]
table_rows=table.select("tbody tr")
columns=table.select("thead th")
columns_names=[c.get_text() for c in columns]
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
pd.DataFrame(l, columns=columns_names)
```

# Grab all the funfacts using Word ''is''

```python
import re
l=[]
funfacts=webpage.select("ul.fun-facts li")

fun=[i.get_text() for i in funfacts]
req=[]
for i in fun:
    if "is" in i:
        req.append(i)
req
```

# Downloading an image from web

```python
img=webpage.find_all("img",attrs={'alt':'Lake Como'})[0]['src']
full_url='https://keithgalli.github.io/web-scraping/'+img
img_data = requests.get(full_url).content
with open('image_name.jpg', 'wb') as handler:
    handler.write(img_data)
```

# SecretWords

```python
sw=webpage.select("div.block a")
for i in sw:
    href=i['href']
    full_url='https://keithgalli.github.io/web-scraping/'+href
    k=requests.get(full_url)
    file=bs(k.content)
    secret_word=file.find_all("p",attrs={'id':'secret-word'})[0].get_text()
    print(secret_word)
    
```
