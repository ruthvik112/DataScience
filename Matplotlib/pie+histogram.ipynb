```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
scores=[100,90,30,20,140,150,20,0,0,100,110]
```

```python
plt.title('Batsman Histogram')
plt.xlabel('Scores')
plt.hist(scores,bins=[0,50,100,150],histtype='barstacked')
```
