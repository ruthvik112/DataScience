```python
company=['GOOGLE','AMAZON','FACEBOOK','MICROSOFT']
revenue=[100,200,70,90]
profit=[100,20,50,70]
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
%matplotlib inline
```

```python
xpos=np.arange(len(company))
```

```python
xpos
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.ylabel('revenue')
plt.xticks(xpos,company)
plt.bar(xpos+0.2,profit,width=0.4,label='profit')
plt.legend(loc="best")
```

```python
plt.barh(xpos-0.2,revenue,label='revenue',height=0.4)
plt.ylabel('revenue')
plt.yticks(xpos,company)
plt.barh(xpos+0.2,profit,label='profit',height=0.4)
plt.legend(loc='best')
plt.grid()
```

```python
plt.plot(xpos,revenue,label="revenue")
plt.xticks(xpos,company)
plt.plot(xpos,profit,label="profit")
plt.title("US Stocks")
plt.legend(loc="best")
plt.grid()
```
